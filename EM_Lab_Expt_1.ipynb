```python
I_f = [0.05, 0.11, 0.21, 0.30, 0.40, 0.50, 0.60, 0.70, 0.80, 0.90, 1.01, 1.1, 1.2]
E_o_f = [30, 57, 101, 132, 158, 180, 195, 207, 220, 230, 239, 245, 253]
E_o_r = [_, 76, 102, 146, 172, 189, 205, 215, 227, 235, 241, 247, 253]

# Check if you entered all the values (above) for all the columns by seeing if each has the same number of values in it
print(len(I_f))
print(len(E_o_f))
print(len(E_o_r))

n = len(I_f)
S_No = list(range(1, n + 1))

df_dict = {
    'S.No.': S_No ,
    'I_f (A)': I_f ,
    'E_o (V) Forward Direction': E_o_f ,
    'E_o (V) Reverse Direction': E_o_r ,
}

df = pd.DataFrame(df_dict)
df.set_index('S.No.', inplace=True)

display(df)
```

# Load Test

```python
V_L = [220, 217, 213, 214, 213, 209, 208, 205]
I_L = [1.88, 3.68, 5.44, 7.21, 8.94, 10.58, 12.22, 13.8]
I_f = [0.79, 0.79, 0.77, 0.77, 0.76, 0.76, 0.74, 0.73]

# Check if you entered all the values (above) for all the columns by seeing if each has the same number of values in it
print(len(V_L))
print(len(I_L))
print(len(I_f))

n = len(I_f)
S_No = list(range(1, n + 1))

df_dict = {
    'S.No.': S_No ,
    'V_L (V)': V_L ,
    'I_L (A)': I_L ,
    'I_f (A)': I_f ,
}

df = pd.DataFrame(df_dict)
df.set_index('S.No.', inplace=True)

R_a = 1.1 # use the required value here!

df['I_a = I_L + I_f'] = df['I_L (A)'] + df['I_f (A)']
df['E = V_L + I_a*R_a'] = df['V_L (V)'] + df['I_a = I_L + I_f'] * R_a
df['N (1500 rpm constant)'] = 500

display(df)
```

```python
import plotly.offline as pyo
pyo.init_notebook_mode(connected=True)
import plotly.graph_objects as go
```

```python
# Create the scatter plot with 'I_a' on the x-axis and 'E' and 'V_L' on the y-axis
scatter = go.Scatter(x=df['I_a = I_L + I_f'], y=df['E = V_L + I_a*R_a'], mode='markers', name='E vs I_a')
scatter_v = go.Scatter(x=df['I_a = I_L + I_f'], y=df['V_L (V)'], mode='markers', name='V_L vs I_a')

layout = go.Layout(
    title={'text': 'E and V_L vs I_a', 'x': 0.5, 'y': 0.95},
    xaxis=dict(title='I_a (A)', titlefont_size=14, tickfont_size=12),
    yaxis=dict(title='E and V_L', titlefont_size=14, tickfont_size=12),
    autosize=False,
    width=600, height=400, margin=dict(l=100, r=50, b=100, t=100, pad=5),
    paper_bgcolor='LightSteelBlue',
    xaxis_showgrid=True,
    yaxis_showgrid=True,
)

fig = go.Figure(data=[scatter, scatter_v], layout=layout)

fig.show(renderer="colab")
```
